```python
cancer['target'] = cancer['target'].apply(lambda x: 0 if x == 2 else 1)
```

```python
fitted_model = smf.logit(formula= 'target ~ sample_code_num + clump_thickness + uniformity_cell_size + uniformity_cell_shape + marginal_adhesion + single_epithelial_cell_size + bland_chromatin + normal_nucleoli + mitoses + bare_nuclei',
data=cancer).fit()
fitted_model.summary()
```
